```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# Define the temperature range
x = np.linspace(20, 150, 100)[:, np.newaxis]
x_pred = np.array([20,60,100,120]).reshape(4,-1)

# Generate synthetic yield data based on a function (you can replace this with your real data)
def true_yield_function(X):
    return (X / 150) * 100  # A simple linear function for demonstration

# Generate synthetic yield data
y = np.array([40,60,90,70]).reshape(4)
print(y.shape)

# Define the Gaussian Process regressor with RBF kernel
kernel = 1.0 * RBF(length_scale=10.0, length_scale_bounds=(1,10))
gp = GaussianProcessRegressor(kernel=kernel, alpha=0.1, normalize_y=True)

# Fit the Gaussian Process model to the data
gp.fit(x_pred, y)

# Make predictions with the trained model
y_pred, y_std = gp.predict(x, return_std=True)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(x_pred, y, color='blue', label='Observations')
plt.plot(x, y_pred, color='red', label='Prediction')
plt.fill_between(x.ravel(), y_pred - y_std, y_pred + y_std, alpha=0.3, color='pink')
plt.title('Temperature', fontsize = 20)
plt.xlabel('Temperature', fontsize = 20)
plt.ylabel('Yield-%', fontsize = 20)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the solvents and their properties
solvents = {
    'Methanol': {'polarity': 1, 'color': 'blue'},
    'Ethanol': {'polarity': 1, 'color': 'green'},
    'Hexane': {'polarity': 0, 'color': 'red'}
}

# Define the latent space
latent_space = np.array([[0, 0], [0.5, 0.5], [2, 2]])

# Plot the solvents in the latent space
plt.figure(figsize=(10, 8))
for solvent, properties in solvents.items():
    plt.scatter(latent_space[:, 0], latent_space[:, 1], c=[properties['color']], label=solvent, s=200)
    plt.text(latent_space[list(solvents.keys()).index(solvent), 0], latent_space[list(solvents.keys()).index(solvent), 1], solvent, fontsize=20)

plt.xlabel('Dimension 1', fontsize=20)
plt.ylabel('Dimension 2', fontsize=20)
plt.title('Solvents in 2D Latent Space', fontsize=20)
plt.legend(prop={'size': 15})
plt.xticks([])
plt.yticks([])
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# Define the temperature range
x = np.linspace(0, 1, 100)[:, np.newaxis]
x_pred = np.array([0.2,0.3, 0.5, 1]).reshape(4,-1)

# Generate synthetic yield data based on a function (you can replace this with your real data)
def true_yield_function(X):
    return (X / 150) * 100  # A simple linear function for demonstration

# Generate synthetic yield data
y = np.array([85,90,70,30]).reshape(4)
print(y.shape)

# Define the Gaussian Process regressor with RBF kernel
kernel = 1.0 * RBF(length_scale=10.0, length_scale_bounds=(0.01,0.1))
gp = GaussianProcessRegressor(kernel=kernel, alpha=0.1, normalize_y=True)

# Fit the Gaussian Process model to the data
gp.fit(x_pred, y)

# Make predictions with the trained model
y_pred, y_std = gp.predict(x, return_std=True)

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(x_pred, y, color='blue', label='Solvent observations')
plt.scatter(0.3, 90, color='Green', label='Mixture of solvent observations')
plt.plot(x, y_pred, color='red', label='Prediction')
plt.fill_between(x.ravel(), y_pred - y_std, y_pred + y_std, alpha=0.3, color='pink', label = 'uncertainty')
plt.title('Optimizing solvents using continuous representations and mixtures', fontsize = 20)
plt.xlabel('Continuous latent representation of solvents', fontsize = 20)
plt.ylabel('Yield-%', fontsize = 20)
plt.legend(prop={'size': 14})
plt.grid(False)
plt.show()
```
